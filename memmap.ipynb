```python
import numpy as np
import tiktoken

#text dataset
texts = [
    "I love machine learning",
    "Large language models are powerful",
    "Neural networks learn patterns"
]

# Loading GPT-2 tokenizer
enc = tiktoken.get_encoding("gpt2")

all_tokens = []

# Tokenize each sentence
for sentence in texts:
    #differentiating the sentences with zero
    all_tokens.append(0)
    ids = enc.encode_ordinary(sentence)

    # saying that this is end of token
    ids.append(enc.eot_token)
    #basically compress into one list rather 
    all_tokens.extend(ids)

print(f"Token IDs:",all_tokens)


#tokenized data as binary file
#the values are 16bit integers
# 1 token = 2 btits
np.array(all_tokens, dtype=np.uint16).tofile("train.bin")


#accesiing binary file using memmap
dataset = np.memmap(
    "train.bin",
    dtype=np.uint16,
    mode="r"
)
```
